```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
benchmarks = pd.read_csv("//content/GPU_benchmarks_v7.csv")
scores = pd.read_csv("/content/GPU_scores_graphicsAPIs.csv")
price_his = pd.read_csv("/content/gpu_price_history.csv")
```

```python
print(benchmarks.head())
print(scores.head())
print(price_his.head())
```

```python
scores.rename(columns={'Device':'gpuName'}, inplace = True)
```

```python
price_his.rename(columns={'Name':'gpuName'}, inplace = True)
```

```python
scores
```

```python
t1 = pd.merge(benchmarks,scores,on = 'gpuName')
```

```python
final = pd.merge(t1,price_his,on = 'gpuName',how = 'left')
```

```python
final
```

```python
final.describe()
```

```python
final.to_csv('output.csv', index=False)
```

```python
final.drop_duplicates()
```

```python
final.isnull().sum()
```

```python
numeric_cols = ['price', 'TDP', 'CUDA', 'gpuValue', 'powerPerformance', 'Metal', 'OpenCL', 'Vulkan']
for col in numeric_cols:
    if col in final.columns:
        final[col] = final[col].fillna(final[col].median())

# Standardize the 'Date' column
final['Date'] = pd.to_datetime(final['Date'], errors='coerce', dayfirst=True)

# Summary statistics
summary_stats = final.describe()
```

```python
col = ['price', 'TDP', 'CUDA', 'gpuValue', 'powerPerformance', 'Metal', 'OpenCL', 'Vulkan']
final[col].hist(bins = 15)
plt.title('Histogram of Numeric Columns')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=final, x='Manufacturer', order=final['Manufacturer'].value_counts().index)
plt.title('Count of GPUs by Manufacturer')
plt.show()
```

```python
#G3 Dmark vs Price
plt.figure(figsize = (10,6))
sns.scatterplot(data = final,x='G3Dmark', y='price', hue='Manufacturer')
plt.title('Performance (G3Dmark) vs. Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=final, x='Manufacturer', y='price')
plt.title('Price Distribution by Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top GPUs visualization
plt.figure(figsize=(12, 8))
sns.boxplot(data=final, x='Price Range', y='G3Dmark', palette='viridis')
plt.title('Performance (G3Dmark) by Price Range')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top GPUs visualization
plt.figure(figsize=(12, 8))
sns.boxplot(data=final, x='Price Range', y='gpuValue', palette='viridis')
plt.title('Performance (G3Dmark) by Price Range')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=final, x='gpuValue', y='price', hue='Manufacturer', size='G3Dmark', sizes=(50, 200))
plt.title('Performance-to-Price vs. Cost by Manufacturer')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=final, x='TDP', y='G3Dmark', hue='Price Range', palette='coolwarm')
plt.title('TDP vs. Performance (G3Dmark)')
plt.show()
```

```python
# Analyze API support by Price Range
api = ['CUDA', 'Metal', 'OpenCL', 'Vulkan']
api_support = final.groupby('Price Range')[api].mean()
plt.figure(figsize=(10, 6))
sns.heatmap(api_support.T, annot=True, fmt='.2f')
plt.title('API Support Levels by Price Range')
plt.show()
```

```python
# Retail Price vs. Used Price Trends
plt.figure(figsize=(12, 6))
sns.scatterplot(data=final, x='Retail Price', y='Used Price', hue='Manufacturer', style='Price Range')
plt.title('Retail Price vs. Used Price Trends by Manufacturer')
plt.show()
```
